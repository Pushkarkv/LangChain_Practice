Data Preprocessing - Chunk Creation (HTMLHeaderTextSplitter)

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_string="""<!DOCTYPE html>
<html>
<head>
    <title>Sample HTML Document</title>
</head>
<body>
    <h1>Introduction</h1>
    <p>Welcome to this sample HTML document. This section introduces the purpose of the document.</p>

    <h2>Overview</h2>
    <p>The overview provides a brief summary of the content discussed here.</p>

    <h3>Purpose</h3>
    <p>The purpose of this document is to provide structured data for testing HTML parsing and text splitting.</p>

    <h3>Scope</h3>
    <p>This document is limited to examples of headers and paragraphs for text processing tasks.</p>

    <h2>Details</h2>
    <p>This section dives into the specific details of the content structure.</p>

    <h3>Subsection 1</h3>
    <p>Subsection 1 includes specific examples to demonstrate hierarchical data organization.</p>

    <h4>Example 1</h4>
    <p>This is an example paragraph under Subsection 1, Example 1.</p>

    <h4>Example 2</h4>
    <p>This is an example paragraph under Subsection 1, Example 2.</p>

    <h3>Subsection 2</h3>
    <p>Subsection 2 provides additional examples and use cases.</p>

    <h2>Conclusion</h2>
    <p>The conclusion summarizes the key points discussed in the document.</p>

    <h3>Summary</h3>
    <p>This document is a simple HTML structure for testing purposes.</p>

    <h3>Next Steps</h3>
    <p>Use this content to test text splitting logic in LangChain and other frameworks.</p>
</body>
</html>
"""
```

```python
headers_to_split_on=[
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3")
]
```

```python
html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
html_header_splits
```

```python
url='https://en.wikipedia.org/wiki/India'

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits=html_splitter.split_text_from_url(url)
html_header_splits
```

Vector Database (FIASS, Ollama)

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader=TextLoader('Testfiles/speech.txt')
documents=loader.load()
text_splitter=CharacterTextSplitter(chunk_size=200, chunk_overlap=30)
docs=text_splitter.split_documents(documents)
```

```python
embeddings=OllamaEmbeddings(model='gemma')
db=FAISS.from_documents(docs,embeddings)
db
```

```python
query="What is the color of Sweetu the crow?"
docs=db.similarity_search(query)
docs[0].page_content
```

Retriever

```python
retriever=db.as_retriever()
retriever.invoke(query)
docs[0].page_content
```

Similarity Search

```python
docs_and_score=db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector=embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=db.asimilarity_search_by_vector(embedding_vector)
docs_score
```

```python
#Saving and Loading the VectorDatabase
db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
docs=new_db.similarity_search(query)
```

ChromaDB

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader=TextLoader('Testfiles/speech.txt')
data=loader.load()
data
```

```python
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
splits=text_splitter.split_documents(data)
```

```python
embeddings=OllamaEmbeddings(model='llama3.1')
vectordb=Chroma.from_documents(splits, embedding=embeddings)
vectordb
```

```python
query="What are the qualities of the crow?"
docs=vectordb.similarity_search(query)
docs[0].page_content
```

```python
#Saving the vectorstore
vectordb=Chroma.from_documents(documents=splits, embedding=embeddings, persist_directory="./chroma_db")
```

```python
db2=Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
```

```python
docs=db2.similarity_search(query)
docs[0].page_content
```

```python
docs=vectordb.similarity_search_with_score(query)
docs
```

```python
retiever=vectordb.as_retriever()
retiever.invoke(query)[0].page_content
```
